## k-space trajectories: ##

```python
def k_x(A=1,dt=0.0001,T=1,n=8,db=0):
    '''
    input:
    A = scaling factor [1/cm]
    dt = time resolution [s]
    T = duration [s]
    n = number of lobes
    '''
    t = np.arange(0,T,dt)
    return A * (1 - t/T)  * np.cos((2*np.pi*n*t)/T)

def k_y(A=1,dt=0.0001,T=1,n=8,db=0):
    '''
    input:
    A = scaling factor [1/cm]
    dt = time resolution [s]
    T = duration [s]
    n = number of lobes
    '''
    t = np.arange(0,T,dt)
    return A * (1 - t/T)  * np.sin((2*np.pi*n*t)/T)

def k_z(A=1,dt=0.0001,T=1):
    '''
    input:
    A = scaling factor [1/cm]
    dt = time resolution [s]
    T = duration [s]
    n = number of lobes
    '''
    t = np.arange(0,T,dt)
    res = A/2-A*t
    return res
```

## Gradients ##

```python
# derivative of kx : Gx = -1 / gamma * dkx/dt
def G_x(nuc='1h',A=1,dt=0.0001,T=1,n=8,gamma=0,db=0):
    '''
    input:
    nuc: nucleus
    A: Amplitude [1/cm]
    dt: time resolution [s]
    T: duration [s]
    gamma: gyromagnetic ratio [MHz/T]
    '''
    if (gamma==0):
        if (nuc == '1h'):
            gamma = 267.52218744 / 2 / np.pi * 1e6
        elif (nuc == '13c'):
            gamma = 67.2828  / 2 / np.pi * 1e6
        else:
            return 0
    t = np.arange(0,T,dt)
    print('using nucleus ([Hz/T])' + str(gamma))
    return -(A/(gamma*T)) * (2 * np.pi * n * (1 - t/T) * np.sin((2 * np.pi * n * t) / T) + np.cos((2 * np.pi * n * t) / T))

# derivative of ky :  Gy = -1 / gamma *  dky/dt
def G_y(nuc='1h',A=1,dt=0.0001,T=1,n=8,gamma=0,db=0):
    '''
    input:
    nuc: nucleus
    A: Amplitude [1/cm]
    dt: time resolution [s]|
    T: duration [s]
    gamma: gyromagnetic ratio [MHz/T]
    '''
    if (gamma==0):
        print('using nucleus info (gamma [Hz/T])')
        if (nuc == '1h'):
            gamma = 267.52218744 / 2 / np.pi * 1e6
        elif (nuc == '13c'):
            gamma = 67.2828 / 2 / np.pi * 1e6
        else:
            return 0
    t = np.arange(0,T,dt)
    return (A/(gamma*T)) * (2 * np.pi * n * (1 - t/T) * np.cos((2 * np.pi * n * t) / T) - np.sin((2 * np.pi * n * t) / T))

# derivative of kz: Gz = -1 / gamma * dkz/dt
def G_z(nuc='1h',A=1,dt=0.0001,T=1,n=8,gamma=0,db=0):
    '''
    input:
    nuc: nucleus
    A: Amplitude [1/cm]
    dt: time resolution [s]|
    T: duration [s]
    gamma: gyromagnetic ratio [MHz/T]
    '''
    if (gamma==0):
        print('using nucleus info (gamma [Hz/T])')
        if (nuc == '1h'):
            gamma = 267.52218744 / 2 / np.pi * 1e6
        elif (nuc == '13c'):
            gamma = 67.2828 / 2 / np.pi * 1e6
        else:
            return 0
    t = np.arange(0,T,dt)
    
    d = np.gradient(k_z(A=A,dt=dt,T=T),t)
    
    res = -1 / gamma * d
    return res
```

## circular Gaussian weighting function ##

```python
def D(a=1,b=1, k_x = np.arange(0,1,0.0001), k_y = np.arange(0,1,0.0001), A=1,db=0):
    '''
    input:
    alpha scales the tipangle
    beta determines the spatial resolution of the excited volume
    k_x, k_y = trajectory through k-space
    return:
    spatial frequency weighting (gaussian)
    '''
    output = a * np.exp(-b*b * (k_x*k_x + k_y*k_y)/(A *A))
    if db:
        print(output.shape)
    return output
```

```python
plt.figure()
plt.plot(D(b=2,a=1, A=3))
plt.title = 'RF Weighting function'
plt.xlabel = 't'
```

## define B1 ##

```python
def B_1(W=np.ones((1,1000)), G_x=np.ones((1,1000)), G_y=np.ones((1,1000)),db=0):
    output = W * np.sqrt(G_x * G_x + G_y * G_y)
    if db:
        print('W=' + str(W.shape))
        print('G_x=' + str(G_x.shape))
        print('G_y=' + str(G_y.shape))
        print('B_1 output shape = ' + str(output.shape))
    return output
```

```python
plt.figure()
plt.subplot(321)
plt.plot(k_x())
plt.subplot(322)
plt.plot(k_y())
plt.subplot(323)
plt.plot(G_x())
plt.subplot(324)
plt.plot(G_y())
plt.subplot(325)
plt.plot(B_1(W=D(b=2,k_x=k_x(),k_y=k_y()), G_x=G_x(), G_y=G_y()))
plt.title = 'B_1'
plt.xlabel = 't [a.u.]'
```

```python
B0 = 1.5 # T 
nuc = '1h'; # protons
Fov = [24, 24] # cm
T = 8 * 1e-3 # s
dt = 0.0000001;
Gmax = 0.6 # G/cm
Gmax_T = 0.6 * 1e-4; # T/cm
n=8
Fov_res = [x / n /2 for x in Fov]
print(Fov_res)
```

```python
%matplotlib inline
gx= G_x(nuc='1h',A=Gmax_T, dt=dt,T=T,n=n)
gy= G_y(nuc='1h',A=Gmax_T, dt=dt,T=T,n=n)
kx= -42577478.517832555 * G_x(nuc='1h',A=Gmax_T, dt=dt,T=T,n=n) * np.exp(1j*np.pi) #np.cumsum(gx) 
ky= -42577478.517832555 * G_y(nuc='1h',A=Gmax_T, dt=dt,T=T,n=n) #* np.cumsum(gy) 



plt.figure()
plt.subplot(121);
plt.plot(gx,gy)
plt.title ='Gradients'
plt.subplot(122);
plt.plot(kx,ky)
plt.title = 'k-space'
print("1/dk = FOV =" + str(1/(2*max(abs(kx))/2/n)))
# (A/gamma*T) * (2 * np.pi * n * (1 - t/T) * np.sin((2 * np.pi * n * t) / T) + np.cos((2 * np.pi * n * t) / T))

# [G/cm] / [Rad/s/T]
```

```python
grads = {'gx' : gx, 'gy': gy}
b1 = B_1(W=D(b=2,k_x=kx,k_y=ky), G_x=gx, G_y=gy,db=1)
b1 = B_1(W=D(b=2,k_x=k_x(),k_y=k_y()), G_x=G_x(), G_y=G_y(),db=1)
b1_plt = B_1(W=D(b=2,
                k_x=k_x(T=T,dt=dt),
                k_y=k_y(T=T,dt=dt)),
                G_x=G_x(T=T,dt=dt),
                G_y=G_y(T=T,dt=dt))
k_xplt = k_x(T=T,dt=dt,db=1);
k_yplt = k_y(T=T,dt=dt,db=1);
plt.plot(np.abs(b1))
b1_dict = {'b1':b1}
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(1, 2, 1, projection='3d')
t = np.arange(0,T,dt)
ax.plot3D(k_xplt,k_yplt,t)
ax.scatter3D(k_xplt,k_yplt,t,c=np.abs(b1_plt)/np.max(np.abs(b1_plt)),marker='.')

ax2 = fig.add_subplot(1, 2, 2, projection='3d')
t = np.arange(0,T,dt)
ax2.plot3D(k_xplt,k_yplt,t)
ax2.view_init(elev=90., azim=180.)
ax2.scatter3D(k_xplt,k_yplt,t,c=np.abs(b1_plt)/np.max(np.abs(b1_plt)),marker='.')
```

#### NUFFT ####

```python
# Initiate the NufftObj object

NufftObj = NUFFT()

# Example from:
# https://jyhmiinlin.github.io/pynufft/index.html#document-tutor/init
import pkg_resources
DATA_PATH = pkg_resources.resource_filename('pynufft', './src/data/')
om = np.load(DATA_PATH+'om2D.npz')['arr_0']
#om = (k_xplt,k_yplt)
#om = np.transpose(om)
#print(np.shape(om))
```

```python
Nd = (256, 256)  # image size
print('setting image dimension Nd...', Nd)
Kd = (512, 512)  # k-space size
print('setting spectrum dimension Kd...', Kd)
Jd = (6, 6)  # interpolation size
print('setting interpolation size Jd...', Jd)

NufftObj.plan(om, Nd, Kd, Jd)


image = np.zeros((256,256))
image[95:150,95:150] = 1;

y = NufftObj.forward(image)

image0 = NufftObj.solve(y, solver='cg',maxiter=50)
image2 = NufftObj.adjoint(y ) # adjoint
image3 = NufftObj.solve(y, solver='L1TVOLS',maxiter=50,rho=0.1)

```

```python
plt.subplot(1,3,1)
plt.title = 'Restored image (cg)'
plt.imshow(np.real(image0));#, cmap=matplotlib.cm.gray, norm=plt.colors.Normalize(vmin=0.0, vmax=1))


plt.subplot(1,3,2)
plt.imshow(np.real(image2));#, cmap=matplotlib.cm.gray, norm=matplotlib.colors.Normalize(vmin=0.0, vmax=5))
plt.title = 'Adjoint transform'


plt.subplot(1,3,3)
plt.title = 'L1TV OLS'
plt.imshow(image3.real);#, cmap=matplotlib.cm.gray, norm=matplotlib.colors.Normalize(vmin=0.0, vmax=1))

plt.show()
```
